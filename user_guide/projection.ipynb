```python
import geodatasets
import geopandas
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql://postgres:postgres@localhost:5432/sdb")
```

```python
usa = geopandas.read_file(geodatasets.get_path('geoda.natregimes'))
usa.crs
```

```python
usa.to_postgis("geod_natregimes_usa", engine)
```

```python
usa.plot()
```

```python
ax = usa.plot()

ax.set_title("WGS84 (lat/lon)");


usa = usa.to_crs("ESRI:102003")

ax = usa.plot()

ax.set_title("NAD 1983 Albers contiguous USA");
```
